# a) Weighted hybrid technique
source: http://trailerpark.weebly.com/imdb-rating.html?source=post_page

# W=(R*v+C*m)/(v+m)
where:
* W=weighted rating 
* R=avg rating at the range (0-10) - "vote_average"
* v= no of votes - "vote_count"
* m= min no. of votes to be considered(here considering 75% of total votes) - "vote_count.quantile(75/100)"
* C=mean of vote across whole data - "vote_average.mean()"

```python
R=data["vote_average"] #initializing the values for the respective
v=data["vote_count"]
C=data["vote_average"].mean()
m=data["vote_count"].quantile(0.75)
```

```python
data["weighted_avg"]=(((R*v)+(C*m))/(v+m)).round(1) #creating a new column "weighted_avg" where the weighted average is stored
```

```python
data.head()
```

```python
data=data.sort_values("weighted_avg",ascending=0) #sorting the data frame with respect to"weighted_avg"
```

```python
weighted_data=data[["original_title","weighted_avg","vote_average","vote_count","popularity"]] 
#selecting the necessary columns to be considered for weighted average
```

```python
weighted_data.head(20) #this dataframe is not considering popularity of the movie so it is biased on weighted average
```

# If we need to consider popularity we shall use 1/2 of weighted avg and 1/2 of popularity to give an unbiased output


```python
scl=MinMaxScaler() #since both weighted_avg and popularity are of different ranges we are using min ax scaler 
data_df=scl.fit_transform(weighted_data[["weighted_avg","popularity"]])
data_norm=pd.DataFrame(data_df,columns=["weighted_avg","popularity"])
data_norm #normalized data
```

```python
data[["Norm_W_Avg","Norm_Popularity"]]=data_norm #appending it to the cleaned data
```

```python
data.head(10)
```

```python
data["rating"]=data["Norm_W_Avg"]*0.5 + data["Norm_Popularity"]*0.5
dt1=data.sort_values("rating",ascending=False)
```

```python
TF=TfidfVectorizer(stop_words="english", #removesfiller words in english sucha s "the,a,and,etc.."
                   min_df=3, #removes terms that appear less than 3 times in our document
                   analyzer="word",#the feature should be made of word n-grams
                   token_pattern=r"\w{1,}",
                   ngram_range=(1,4),#the range of n-values for different n-grams to be extracted
                   strip_accents="unicode")#removes the accents since differnt accents may confuse the machine
data["overview"]=data["overview"].fillna('')
```

```python
TF_mat=TF.fit_transform(data["overview"]) #fitting the values to form a matrix
```

```python
TF_mat.shape #this matrix includes the vectorized form of words in overview
```

```python
sig=sigmoid_kernel(TF_mat,TF_mat)
```
